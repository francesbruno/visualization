```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

Basic figure w/ matplotlib

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50)
```

basic scatterplot


```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)
```

```python
#edit axis object
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)
```

```python
#line plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
```

```python
#histogram
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

Adding labels to scatterplot and histogram overlays

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
ax.scatter(x,y)
ax.set_title('Total Growth over Time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

```python
#formatting
font1 = {'family':'sans-serif'
,
'color':'blue'
,
'size':20}
font2 = {'family':'monospace'
,
'color':'green'
,
'size':14}
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time'
, fontdict = font1)
ax.set_ylabel('Total growth'
, fontdict = font2)
ax.set_xlabel('Years since start'
, fontdict = font2)
fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y,marker='*', # NEW ADDITIONS
color = "indigo") # NEW ADDITIONS
fig.show()
```

```python
#modifying data points
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*',color = 'indigo',linestyle = '--', # NEW ADDITIONS
linewidth = 2) # NEW ADDITIONS
fig.show()
```

```python
#changing color using hex code
fig, ax = plt.subplots(figsize=(5, 3))
# color being the new addition
ax.plot(x,y,marker='*', color = '#7425b9', linestyle = '--', linewidth = 2)
fig.show()
```

```python
#modifying data points

fig, ax = plt.subplots(figsize=(5, 3))
# color being the new addition
ax.plot(x,y,marker='*', color = '#7425b9', linestyle = '--', linewidth = 2
, markeredgecolor = '#fa9359', # NEW ADDITIONS
markerfacecolor = '#000000')  # NEW ADDITIONS 
fig.show()
```

```python
# + adding axis

fig, ax = plt.subplots(figsize=(5, 3))
# color being the new addition
ax.plot(x,y,marker='*', color = '#7425b9', linestyle = '--', linewidth = 2
, markeredgecolor = '#fa9359', # NEW ADDITIONS
markerfacecolor = '#000000')  # NEW ADDITIONS 
fig.show()
ax.grid(axis = 'y')
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.grid(axis = 'y', color = "blue", linewidth = 2, linestyle = '-.')
ax.plot(x,y,marker='*',markersize = 12, color = '#7425b9',linestyle = '--',linewidth = 2, markeredgecolor = '#fa9359', # NEW ADDITIONS
markerfacecolor = '#000000' )
```
